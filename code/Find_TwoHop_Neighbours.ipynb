```python
import numpy as np
import pandas as pd
import dask
import time
from scipy.sparse import coo_matrix, find

np.random.seed(1)

# Using readlines()
file1 = open('./com-dblp.ungraph.txt', 'r')
Lines = file1.readlines()
connected = []
 
# Strips the newline character
for line in Lines:
    a = line.split()
    results = list(map(int, a))
    connected.append(results)
    
original = np.asarray(connected)
rng = len(np.unique(original))
vertex = np.unique(original)
mapping = dict(map(lambda t: (t[1], t[0]), enumerate(vertex)))
iterator = map(lambda arr: [mapping[arr[0]],mapping[arr[1]]], original)
gr = np.asarray(list(iterator))

grsp = coo_matrix((np.ones(gr.shape[0]), (gr[:, 0], gr[:, 1])),
                      shape=(rng, rng))
grcsr = grsp.tocsr()
adj = grcsr**2
result = find(adj)
```

```python
result
```

```python
t = [[vertex[a],vertex[b]] for a, b in zip(result[0], result[1])]
```

```python
t.sort(key=lambda arr: arr[0])
```

```python
f = open("2hop_final.txt", "a")
for row in t:
    f.write(str(row[0]))
    f.write(' ')
    f.write(str(row[1]))
    f.write('\n')
f.close()
```
